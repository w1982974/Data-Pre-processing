```python
import pandas as pd
import numpy as np
from sklearn.utils import shuffle
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('Edge-IIoTset dataset\\Selected dataset for ML and DL\\ML-EdgeIIoT-dataset.csv', low_memory=False) 
df.head(5)
```

```python
print(df['Attack_type'].value_counts())
```

# Pre-processin

```python
drop_columns = ["frame.time", "ip.src_host", "ip.dst_host", "arp.src.proto_ipv4","arp.dst.proto_ipv4", 
                "http.file_data","http.request.full_uri","icmp.transmit_timestamp",
                "http.request.uri.query", "tcp.options","tcp.payload","tcp.srcport",
                "tcp.dstport", "udp.port", "mqtt.msg"]

df.drop(drop_columns, axis=1, inplace=True)
df.dropna(axis=0, how='any', inplace=True)
df.drop_duplicates(subset=None, keep="first", inplace=True)
df = shuffle(df)
df.isna().sum()
```

```python
print(df['Attack_type'].value_counts())
```

```python
dummies = pd.get_dummies(df['http.request.method'],prefix=f"'http.request.method", drop_first=True, dtype=int)
df = pd.concat([df,dummies], axis=1)
df.drop('http.request.method', axis=1, inplace=True)

dummies = pd.get_dummies(df["http.referer"],prefix="http.referer",drop_first=True, dtype=int)
df = pd.concat([df,dummies], axis=1)
df.drop("http.referer", axis=1, inplace=True)

dummies = pd.get_dummies(df["http.request.version"],prefix="http.request.version",drop_first=True, dtype=int)
df = pd.concat([df,dummies], axis=1)
df.drop("http.request.version", axis=1, inplace=True)

dummies = pd.get_dummies(df["dns.qry.name.len"],prefix="dns.qry.name.len",drop_first=True, dtype=int)
df = pd.concat([df,dummies], axis=1)
df.drop("dns.qry.name.len", axis=1, inplace=True)

dummies = pd.get_dummies(df["mqtt.conack.flags"],prefix="mqtt.conack.flags",drop_first=True, dtype=int)
df = pd.concat([df,dummies], axis=1)
df.drop("mqtt.conack.flags", axis=1, inplace=True)

dummies = pd.get_dummies(df["mqtt.protoname"],prefix="mqtt.protoname",drop_first=True, dtype=int)
df = pd.concat([df,dummies], axis=1)
df.drop("mqtt.protoname", axis=1, inplace=True)

dummies = pd.get_dummies(df["mqtt.topic"],prefix="mqtt.topic",drop_first=True, dtype=int)
df = pd.concat([df,dummies], axis=1)
df.drop("mqtt.topic", axis=1, inplace=True)
```
